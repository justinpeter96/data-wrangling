# Occupation

### Step 1. Import the necessary libraries

```python
import pandas as pd
import numpy as np
```

### Step 2. Import the dataset from this [address](https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user). 

### Step 3. Assign it to a variable called users.

```python
url = 'https://raw.githubusercontent.com/justmarkham/DAT8/master/data/u.user'
users = pd.read_csv(url, sep='|')
```

### Step 4. Discover what is the mean age per occupation

```python
users = users.loc[:,'age':'zip_code']
users.groupby('occupation').mean('age')
```

### Step 5. Discover the Male ratio per occupation and sort it from the most to the least

We delete a couple columns and adjust dataset by the male gender

```python
maleusers = users.loc[users.gender == 'M']
del maleusers['age']
del maleusers['zip_code']
maleusers.head()
```

```python
maleusers.groupby(['occupation', 'gender']).size().sort_values(ascending=False)
```

### Step 6. For each occupation, calculate the minimum and maximum ages

```python
describe = users.groupby('occupation').describe()
describe.iloc[:,np.r_[3,7]]
```

### Step 7. For each combination of occupation and gender, calculate the mean age

```python
users.groupby(['occupation','gender']).mean('age')
```

### Step 8.  For each occupation present the percentage of women and men

First line displays count for both genders on each occupation

```python
del users['zip_code']
users=users.rename(columns = {'age':'count'})
users.groupby(['occupation','gender']).count() # count for both genders
```

Second line displays sum of counts for each occupation

```python
users.groupby(['occupation'])['gender'].count() # sum up gender
```

Final line produces the percentage

```python
gender = users.groupby(['occupation','gender'])['gender'].count()
total = users.groupby(['occupation'])['gender'].count() 
print(gender/total)
```
