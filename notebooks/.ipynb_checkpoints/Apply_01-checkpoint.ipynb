### Step 5. Create a lambda function that capitalize strings.

```python
capitalize = lambda x: x.upper()
```

### Step 6. Capitalize both Mjob and Fjob

```python
Mjob = df['Mjob'].apply(capitalize)
Fjob = df['Fjob'].apply(capitalize)
print(Mjob)
print(Fjob)
```

### Step 7. Print the last elements of the data set.

```python
print(df.iloc[-1])
```

### Step 8. Did you notice the original dataframe is still lowercase? Why is that? Fix it and capitalize Mjob and Fjob.

```python
d = {'Mjob': Mjob, 'Fjob': Fjob}
df1 = pd.DataFrame(data=d)
```

```python
df = df.assign(Mjob=df1['Mjob'])
df = df.assign(Fjob=df1['Fjob'])
print(df.iloc[-1])
```

### Step 9. Create a function called majority that returns a boolean value to a new column called legal_drinker (Consider majority as older than 17 years old)

```python
majority = df['age'] > 17
```

```python
df['legal_drinker'] = majority
```
